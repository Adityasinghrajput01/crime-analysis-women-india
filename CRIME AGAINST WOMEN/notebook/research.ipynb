```python
X = df[[
    "Rape",
    "Kidnapping & Abduction_Total",
    "Cruelty by Husband or his Relatives",
    "Assault on Women with intent to outrage her Modesty_Total"
]]

y = df["Total Crimes against Women"]
```

```python
# Since the target variable is continuous (total crime count), this is a supervised regression problem.
```

```python
from sklearn.preprocessing import StandardScaler

# Creating scaler object
scaler = StandardScaler()

# Scaling the features (important for ML models)
X_scaled = scaler.fit_transform(X)

# X_scaled = normalized version of the features
```

```python
from sklearn.model_selection import train_test_split

# Spliting data into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict on test data
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# MAE (average error)
mae = mean_absolute_error(y_test, y_pred)

# MSE (squared error)
mse = mean_squared_error(y_test, y_pred)

# R2 Score (accuracy for regression)
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("R2 Score:", r2)
```

```python
# R2 Score (MOST IMPORTANT)
#  0.965 (≈ 96.5%)

#  This means:

# Model explains 96.5% of the variance
# Very high accuracy

# Interpretation :

# "The model performs very well and captures most of the relationship between features and target."
```
